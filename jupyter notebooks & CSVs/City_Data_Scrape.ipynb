```python
def df_to_geojson(city_df, properties, lat='latitude', lon='longitude'):
    geojson = []
    for _, row in city_df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson.append(feature)
    return geojson
```

```python
city_cols = ["city","state","Median_Per_Capita_Income", "Median_Rent", "Median Monthly Income"]
```

```python
geojson=df_to_geojson(city_df,city_cols)
```

```python
output_filename = 'geocities.js'
with open(output_filename, 'w') as output_file:
    output_file.write('var city_data = ')
    json.dump(geojson, output_file, indent=2) 
```
