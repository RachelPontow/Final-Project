```python
import pandas as pd
import json

```

```python
college_data = pd.read_json("college.json")
```

```python
college_data = college_data.rename(columns= {"Debt for Income <30K":"Debt_Less_Thirty", "Debt for Income between 31K & 75K":"Debt_Less_Seventyfive", "Debt for Income greater than 75K":"Debt_Greater_Seventyfive", "Monthly Payments": "Monthly_Payments" })
```

```python
college_data
```

```python
college_data.dtypes
```

```python
college_data=college_data.round({"MonthlyPayments":1})
college_data.head()
```

```python
college_data.to_csv("final_college.csv", index = False)
```

```python
def df_to_js(college_data, properties):
    college = []
    for _, row in college_data.iterrows():
        feature = {}

        for prop in properties:
            feature[prop] = row[prop]
        college.append(feature)
    return college
```

```python
cols = ["Name", "Monthly_Payments", "Debt_Less_Thirty", "Debt_Less_Seventyfive", "Debt_Greater_Seventyfive"]
```

```python
college=df_to_js(college_data,cols)
```

```python
print(college)
```

```python
output_filename = 'collegefinal.js'
with open(output_filename, 'w') as output_file:
    output_file.write('var college_info = ')
    json.dump(college, output_file, indent=2) 
```

```python
city_data = pd.read_json('cities2.json')
```

```python
city_data= city_data.rename(columns = {"Median Monthly Income": "Median_Monthly_Income"})
```

```python
city_data= city_data[["Median_Monthly_Income", "Median_Rent", "city", "state"]]
```

```python
city_data.head()
```

```python
city_data.to_csv("final_city.csv", index = False)
```

```python
def df_to_js2(city_data, properties):
    city = []
    for _, row in city_data.iterrows():
        feature = {}

        for prop in properties:
            feature[prop] = row[prop]
        city.append(feature)
    return city
```

```python
cols2=["Median_Monthly_Income", "Median_Rent", "city", "state"]
```

```python
city=df_to_js2(city_data,cols2)
```

```python
print(city)
```

```python
output_filename = 'citiesfinal.js'
with open(output_filename, 'w') as output_file:
    output_file.write('var city_info = ')
    json.dump(city, output_file, indent=2) 
```
