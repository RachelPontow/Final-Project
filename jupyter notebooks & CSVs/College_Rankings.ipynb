```python
id_df=pd.read_csv('Most-Recent-Cohorts-Scorecard-Elements.csv')
```

```python
id_df
```

```python
id_df=id_df[["INSTNM", "UNITID", "CITY", "STABBR"]]
id_df
```

```python
remove_words = ['The ']
pat = r'\b(?:{})\b'.format('|'.join(remove_words))

id_df["INSTNM"] = id_df["INSTNM"].str.replace(pat, '')
```

```python
id_df=id_df.set_index(["INSTNM"])

```

```python
the_final_df = pd.merge(df, id_df, left_index=True, right_index=True)
```

```python
the_final_df
```

```python
id_list = the_final_df["UNITID"].tolist()
```

```python
print(id_list)
```

```python
name = []
lat = []
lon = []

for item in id_list:
    url = "https://api.data.gov/ed/collegescorecard//v1/schools.json?"
    query = f"{url}&api_key={api_key}&id={item}"
    response = requests.get(query).json()
    i = 0
    for x in response['results']:
        name.append(response['results'][i]['school']['name'])
        lat.append(response['results'][i]['location']['lat'])
        lon.append(response['results'][i]['location']['lon'])
        
        i+=1
    
    
```

```python
print(lon)
```

```python
college_df = pd.DataFrame({
    "Name": name,
    "Latitude": lat,
    "Longitude": lon
})
```

```python
college_df
```

```python
college_df.to_json("college.json", orient="records")
```

```python
name2 = []
debt_less_thirty = []
debt_less_seventyfive = []
debt_greater_seventyfive = []
monthly_pay = []

for item in id_list:
    url = "https://api.data.gov/ed/collegescorecard//v1/schools.json?"
    query = f"{url}&api_key={api_key}&id={item}"
    response = requests.get(query).json()
    i = 0
    for x in response['results']:
        name2.append(response['results'][i]['school']['name'])
        debt_less_thirty.append(response['results'][i]['2015']['aid']['median_debt']['income']['0_30000'])
        debt_less_seventyfive.append(response['results'][i]['2015']['aid']['median_debt']['income']['30001_75000'])
        debt_greater_seventyfive.append(response['results'][i]['2015']['aid']['median_debt']['income']['greater_than_75000'])
        monthly_pay.append(response['results'][i]['2015']['aid']['median_debt_suppressed']['completers']['monthly_payments'])
        
        
        i+=1
    
```
